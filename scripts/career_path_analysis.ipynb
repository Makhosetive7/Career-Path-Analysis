01: Salary Insights analysis

```python

#Average salary across all job  titles paid in USD
average_USD_salary = dataFrame["salary_in_usd"].mean()
print(f"Average Salary: ${average_USD_salary:.2f}")
```

1. Which data roles offer the highest salaries for entry-level positions?

```python

entry_level = dataFrame[dataFrame['experience_level'] == 'Entry-level']
plt.figure(figsize=(20,6))
sns.barplot(data=entry_level, x='job_title', y='salary_in_usd', estimator=sum)
plt.xticks(rotation=45)
plt.title("Average Salary by Job Title (Entry-level)")
plt.show()
```

2. How do salaries differ between mid-level and senior positions across data roles?

```python
plt.figure(figsize=(20,6))
sns.boxplot(data=dataFrame[dataFrame['experience_level'].isin(['Mid-level','Senior'])],
            x='job_category', y='salary_in_usd', hue='experience_level')
plt.xticks(rotation=45)
plt.title("Salary Distribution: Mid-level vs Senior by Job Category")
plt.show()
```

3. Does working remotely, hybrid, or in-person affect salary?

```python
plt.figure(figsize=(20,6))
sns.violinplot(data=dataFrame, x='work_setting', y='salary_in_usd', palette='Set2')
plt.title("Salary Distribution by Work Setting")
plt.show()
```

4. Which data job categories pay best on average?

```python
avg_salary = dataFrame.groupby('job_category')['salary_in_usd'].mean().sort_values()
avg_salary.plot(kind='barh', figsize=(20,6), color='skyblue')
plt.xlabel("Average Salary (USD)")
plt.title("Average Salary by Job Category")
plt.show()
```

5. Do larger companies pay more than smaller companies?

```python
plt.figure(figsize=(20,6))
sns.boxplot(data=dataFrame, x='company_size', y='salary_in_usd', palette='pastel')
plt.title("Salary Distribution by Company Size")
plt.show()
```

6. Which countries/cities offer highest salaries?

```python
top_countries = dataFrame.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
top_countries.plot(kind='bar', figsize=(20,6), color='coral')
plt.ylabel("Average Salary (USD)")
plt.title("Top 10 Countries by Average Salary")
plt.show()
```

7. How does experience level influence salary growth?

```python
average_by_experience = dataFrame.groupby('experience_level')['salary_in_usd'].mean()
average_by_experience.plot(kind='line', marker='o', figsize=(20,6))
plt.ylabel("Average Salary (USD)")
plt.title("Salary vs Experience Level")
plt.show()
```

8. Which combination of role, location, and work setting maximizes salary?

```python
pivot = dataFrame.pivot_table(values='salary_in_usd', index='job_category', columns='work_setting', aggfunc='mean')
plt.figure(figsize=(20,6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title("Average Salary by Job Category & Work Setting")
plt.show()
```

9. Roles with fastest salary growth

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data=dataFrame, x='experience_level', y='salary_in_usd', hue='job_category', style='job_category', s=100)
plt.title("Salary Growth Across Experience Levels by Job Category")
plt.show()
```
